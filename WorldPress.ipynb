```python
import requests
from bs4 import BeautifulSoup
```

```python
import pandas as pd
```

```python
URL = 'https://rsf.org/en/ranking_table'
```

```python
page = requests.get(URL)
```

```python
soup = BeautifulSoup(page.text, 'html5lib')
table = soup.find('table')
```

```python
data = []
for row in table.find_all('tr')[1:]:
    data.append([])
    for i, column in enumerate(row.find_all('td')):
        if i == 0:
            data[-1].append(column['style'].replace('background:', ''))
        elif i != 2:
            data[-1].append(float(column.text))
        else:
            data[-1].append(column.text)
```

```python
df = pd.DataFrame(data)
```

```python
import requests
import io
import pandas as pd

response = requests.get('https://rsf.org/sites/default/files/index_2019_-_pour_import_1_1.csv')

with io.StringIO(response.content.decode('utf-8')) as f:
    df = pd.read_csv(f, decimal=',')
```

```python
df.sort_values('Score 2019').tail(50)
```

```python
DICT_MATCHING = {
    'Norway': 'NOR',
    'France': 'FRA',
    'Somaliland': 'SOM',
    'N. Cyprus' :'CTU',
    'Kosovo': 'XKO'
}
```

```python
import cartopy.crs as ccrs
import matplotlib.pyplot as plt
import cartopy.io.shapereader as shpreader
import numpy as np

plt.rcParams['figure.figsize'] = [15, 8]
shapename = 'admin_0_countries'
countries_shp = shpreader.natural_earth(
    resolution='110m',
    category='cultural',
    name=shapename
)

ax = plt.axes(projection=ccrs.PlateCarree())

cmap = plt.get_cmap('afmhot_r')
max_val = df['Score 2019'].max()

for country in shpreader.Reader(countries_shp).records():
    name_country = country.attributes['ISO_A3']
    name_country = DICT_MATCHING.get(country.attributes['NAME'], name_country)
    if name_country in df['ISO'].values:
        val = df[df['ISO'] == name_country]['Score 2019'].values[0]
        color = cmap([val / max_val + 0.15])
    else:
        color = '#ffffff'

    ax.add_geometries(
        country.geometry, 
        ccrs.PlateCarree(),
        facecolor=color
    )
```
